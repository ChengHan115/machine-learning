# **第二題的第二，三小題**

```python
from google.colab import drive
drive.mount("/content/drive")
import pandas as pd

import numpy as np
from sklearn.linear_model import LinearRegression
df = pd.read_csv("/content/drive/MyDrive/機器學習/作業5/BMI-5_9.csv")

# 提取1975-2016年的數據
df = df[(df['Period'] >= 1975) & (df['Period'] <= 2016)]

# 提取需要的特徵和標籤
X = df[['Period']]
y = df[['Mean BMI']]

# 建立並訓練線性迴歸模型
lr = LinearRegression()
lr.fit(X, y)

X_2025 = [[2025]]
y_pred_2025 = lr.predict(X_2025)
# print('2025年平均BMI預測值為：{:.2f}'.format(y_pred_2025[0][0]))
bmi_2025 = y_pred_2025[0][0]
print(bmi_2025)

bmi_2000 = 15.15  # 2000年125cm身高的BMI平均值
bmi_diff = bmi_2025 - bmi_2000
height = 1.25  
weight_diff = bmi_diff * height**2  # 計算體重差異
print("2025年125cm的兒童體重比2000年同樣身高的兒童提高了{:.2f}kg".format(weight_diff))
```
