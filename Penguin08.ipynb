# **第一題的第二小題**

```python
from google.colab import drive
drive.mount("/content/drive")
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/機器學習/作業5/1880-2020年均溫.csv")
# df.plot(kind='scatter',x='Decade',y='Change from 1880')

import numpy as np
from sklearn.linear_model import LinearRegression

# 提取需要的特徵和標籤
X = df[['Decade']]
y = df[['Change from 1880']]

# 建立並訓練線性迴歸模型
lr = LinearRegression()
lr.fit(X, y)

# 預測2030-2040十年均溫趨勢
X_2030_2040 = np.array([[2030]])
y_pred_2030_2040 = lr.predict(X_2030_2040)

# 預測2050-2060十年均溫趨勢
X_2050_2060 = np.array([[2050]])
y_pred_2050_2060 = lr.predict(X_2050_2060)

print('2030-2040十年均温趨勢预测:', y_pred_2030_2040[0][0])
print('2050-2060十年均温趨勢预测:', y_pred_2050_2060[0][0])
```

# **第一題的第四小題**

```python
from google.colab import drive
drive.mount("/content/drive")
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/機器學習/作業5/1880-2020年均溫.csv")
# df.plot(kind='scatter',x='Decade',y='Change from 1880')

import numpy as np
from sklearn.linear_model import LinearRegression

# 提取1970-2020年的數據
df = df[(df['Decade'] >= 1970) & (df['Decade'] <= 2020)]

# 提取需要的特徵和標籤
X = df[['Decade']]
y = df[['Change from 1880']]

# 建立並訓練線性迴歸模型
lr = LinearRegression()
lr.fit(X, y)

# 預測2030-2040十年均溫趨勢
X_2030_2040 = np.array([[2030]])
y_pred_2030_2040 = lr.predict(X_2030_2040)

# 預測2050-2060十年均溫趨勢
X_2050_2060 = np.array([[2050]])
y_pred_2050_2060 = lr.predict(X_2050_2060)

print('2030-2040十年均温趨勢预测:', y_pred_2030_2040[0][0])
print('2050-2060十年均温趨勢预测:', y_pred_2050_2060[0][0])
```
